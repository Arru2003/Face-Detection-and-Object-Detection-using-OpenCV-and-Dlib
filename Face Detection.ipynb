```python

# Load image
image = cv2.imread('C:\\Users\\aruni\\Downloads\\Data Set\\people1.jpg')

# Check if the image was successfully loaded
if image is None:
    print("Image not found or unable to read.")
else:
    # Resize image
    image = cv2.resize(image, (800, 600))

    # Convert image to grayscale
    image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Detect faces
    detections = face_detector.detectMultiScale(image_gray, scaleFactor=1.09)

    # Draw rectangles around detected faces
    for (x, y, w, h) in detections:
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 5)

    # Display the image with rectangles drawn around faces
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Turn off axis
    plt.show()
```

```python

# Load image
image = cv2.imread('C:\\Users\\aruni\\Downloads\\Data Set\\people3.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Load Haar cascade classifier
haarcascade_detector = cv2.CascadeClassifier('C:\\Users\\aruni\\Downloads\\Cascades\\haarcascade_frontalface_default.xml')

# Detect faces
detections = haarcascade_detector.detectMultiScale(image_gray, scaleFactor=1.001, minNeighbors=5, minSize=(5, 5))

# Draw rectangles around detected faces
for (x, y, w, h) in detections:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Display the image with rectangles drawn around faces
cv2.imshow('', image)
cv2.waitKey(10000)
plt.show()
cv2.destroyAllWindows()
```
